# Analyse du fond
### 1. Supprimer les colonnes inutiles DATA CLEANING
 

```python
# supprimer la colonne Patient Id
df = data.drop('Patient Id',axis=1)
df.head()
```

### 2. Analyser la variable target

```python
df['Level'].value_counts(normalize=True)
```

On remarque que nos class sont équilibré
sur 1000 patients nous avons 36% qui ont un niveau de cancer avancé, 33% qui ont un niveau de cancer moyen et 30% qui ont un niveau de cancer léger.

# Histogrames des variables

### 1. Visualisation des variables quantitatives

```python
plt.figure(figsize = (20,10))
for col in df.select_dtypes('int'):
    sns.displot(df[col], kde=True)
```

### On remarque ici qu'un nombre important d'individu de notre dataset ont : 
 - ##### une obésité élévé
 - ##### ont un risque  gènes
 - ##### risque professionel
 - ##### allergie à la poussiére
 - ##### Coughing of blood (le faite de cracher du sang)

### 2. Visualisation de la variable target


```python
df['Level'].value_counts().plot.pie()
```

##### Comme mentionné précédement nous constatons que le pourcentage de nos class est équilibré 

### 3- Encoder notre variable target

```python
df.loc[:,'Level'] = df['Level'].map({'Low':0,'Medium':1,'High':2})
df['Level'].unique()
```

# Relation entre notre variable target et les autres variables


# Target/Age

```python
print(f"Ces analyses ont été collecter sur des personnes âgées entre : {df['Age'].min()} ans et {df['Age'].max()} ans ce qui fait une moyenne d'âge : {df['Age'].mean()} ans")
```

```python
fig, ax=plt.subplots()

plot = sns.boxplot(data=df, x='Level', y='Age', palette=['darkblue','darkred','darkgreen'])

plt.xlabel("Level",   color='k')
plt.ylabel("Age",  color='k')
plot
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x = 'Age',hue='Level',data=df);
```

On remarque qu'il n'y a pas d'âge précis pour attraper un cancer par contre la tranche d'âge 33 à 38 ans présente plus de risque d'avoir un cancer.
Nous n'avons pas de données dans ce dataset concernant les individus âgés entre 0 et 14 ans ce qui ne veut pas dire qu'ils ne sont pas exposé au risque de cancer.
