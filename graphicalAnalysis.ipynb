# Analyse du fond
### 1. Supprimer les colonnes inutiles DATA CLEANING
 

```python
# supprimer la colonne Patient Id
df = data.drop('Patient Id',axis=1)
df.head()
```

### 2. Analyser et Visualiser la variable target

```python
df['Level'].value_counts(normalize=True)
```

```python
df['Level'].value_counts().plot.pie()
```

On remarque que nos class sont équilibré
sur 1000 patients nous avons 36% qui ont un niveau de cancer élevé, 33% qui ont un niveau de cancer moyen et 30% qui ont un niveau de cancer léger.

### 3- Encoder notre variable target

```python
df.loc[:,'Level'] = df['Level'].map({'Low':0,'Medium':1,'High':2})
df['Level'].unique()
```

# Target/Age

```python
print(f"Ces analyses ont été collecter sur des personnes âgées entre : {df['Age'].min()} ans et {df['Age'].max()} ans ce qui fait une moyenne d'âge: {df['Age'].mean()} ans")
```

```python
fig, ax=plt.subplots()

plot = sns.boxplot(data=df, x='Level', y='Age')

plt.xlabel("Level")
plt.ylabel("Age")
plot
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x = 'Age',hue='Level',data=df);
```

On remarque ici que pour les trois niveau de cancer la majorité des individus ont entre 30 et 45 ans et je ne pense pas que l'age joue un role dans le niveau de cancer
On remarque aussi que nous avans une valeurs abberante qui peut fausser la moyenne d'age pour les individus qui ont un cancer de niveau Medium car la valeur est plus élevée que les autres valeurs dans cette class
Nous n'avons pas de données dans ce dataset concernant les individus âgés entre 0 et 14 ans ce qui ne veut pas dire qu'ils ne sont pas exposé au risque de cancer.  

```python
import math
i,j=0,0;
PLOTS_PER_ROW = 5; # afficher 5 boxplots par ligne
fig, axes = plt.subplots(math.ceil(len(df.columns)/PLOTS_PER_ROW),PLOTS_PER_ROW, figsize=(15,15));
for col in df.columns:
    sns.boxplot(ax=axes[i, j], data=df, x='Level', y=df[col]);
    axes[i][j].set_ylabel(col);
    j+=1;
    if j%PLOTS_PER_ROW==0:# si j modulo plots per row = 0 je reinitialise le compteur j et j'incrémente mon i qui représente un row
        i+=1;
        j=0;
plt.show();
```

```python
fig = plt.figure(figsize = (20,20))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True);
```
